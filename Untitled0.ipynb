```python
import cudf
import numpy as np

df = cudf.read_csv('/content/Google-Playstore.csv')
```

```python
df.info()
```

```python
df["Ad Supported"]
```

```python
df.drop(columns=["Developer Website","Developer Email","Rating Count","Scraped Time","Rating Count"],inplace=True)
```

```python
df["Minimum Installs"] = df["Minimum Installs"].astype(np.int8)
```

```python
df["Maximum Installs"] = df["Maximum Installs"].astype(np.int8)
```

```python
bool_columns = df.select_dtypes("bool").columns

for col in bool_columns:
    df[col] = df[col].astype(np.int8)
```

```python
df.drop(columns=["Privacy Policy"],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

object_columns = df.select_dtypes("object").columns

for col in object_columns[4 :- 4]:
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col].to_numpy()).astype(np.int16)
```

```python
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
```

```python

avg_installs = df.groupby("Ad Supported")["Maximum Installs"].mean()
avg_installs = avg_installs.to_pandas()

mean = avg_installs.mean()
std = avg_installs.std()

plt.hlines(mean,1,10,color="g", linestyle="--")
plt.hlines(mean+std*2,1,10,color="r", linestyle="--")
plt.hlines(mean-std*2,1,10,color="r", linestyle="--")

avg_installs.plot.bar()
plt.title("Ads vs No Ads")
plt.ylabel("Average Installs")
plt.xticks([0,1], ["No Ads", "Ads"])
plt.show()
```

```python

plt.scatter(df["Price"].to_numpy(), df["Rating"].to_numpy(), alpha=0.5)
mean = df["Price"].mean()
std = df["Price"].std()
plt.hlines(mean,1,10,color="g", linestyle="--")
plt.hlines(mean+std*2,1,10,color="r", linestyle="--")
plt.hlines(mean-std*2,1,10,color="r", linestyle="--")
plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

```python
top_categories = df.groupby("Category")["Maximum Installs"].mean().head(10)
top_categories_pandas = top_categories.to_pandas()
top_categories_pandas.plot.bar()

mean = top_categories_pandas.mean()
std = top_categories_pandas.std()

plt.hlines(mean,0,9,color="g", linestyle="--")
plt.hlines(mean+std*2,0,9,color="r", linestyle="--")
plt.hlines(mean-std*2,0,9,color="r", linestyle="--")

plt.title("Top Categories")
plt.xlabel("Categories")
plt.ylabel("Installs")
plt.show()
```

```python
avg_rating = df.groupby("Free")["Rating"].mean()
avg_rating= avg_rating.to_pandas()

mean = avg_rating.mean()
std = avg_rating.std()

plt.hlines(mean,0,10,color="g", linestyle="--")
plt.hlines(mean+std*2,0,10,color="r", linestyle="--")
plt.hlines(mean-std*2,0,10,color="r", linestyle="--")

avg_rating.plot.bar()
plt.title(" Free vs Paid Apps")
plt.ylabel("Average Rating")
plt.xlabel("App Type")
plt.xticks([0,1], ["Paid", "Free"])
plt.show()
```
